<pre>Problem Statement 1: In a survey conducted by a non-banking financial company, a sample of 200 customers yielded that x of them were highly satisfied with the timely disbursal of their loans.Write a Python code to perform the following operations:1.Read an integer input thatspecifies the number of highly satisfied customers2.Calculate an approximate 90% confidence interval for the proportion of the loan customers who are highly satisfied with disbursal time•Find out the Margin of Error using scipy.stats.norm.ppf•Calculate and print the confidence interval values rounded up to five decimal places and separated by a space</pre>



```python
import numpy as np
import scipy.stats as st
from math import sqrt 
n = 200
x = int(input("Please Enter a number of happy customers"))
p = x/n
#critical value
critical_value = st.norm.ppf(.90)
se = sqrt((p*(1-p))/n)
# Marginal Error = critical_value*se
margin_of_error=critical_value*se
# CI
# (p-margin) , (p+margin)
print("the cofidence interval is:",np.round((p-margin_of_error,p+margin_of_error),5))
```

Problem Statement 4: Write a Python program to load the wine dataset using the Sklearn 
library to a DataFrame and perform the following tasks:
1. Convert the dataset into DataFrame using pandas.
2. Generate the sample size of 50 and give a random state as 100.
3. Calculate Z-critical, Margin of Error, and Confidence Interval for alcohol at 95% 
significance interval on generated sample data.

```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
wine = load_wine()
X = wine.data
y = wine.target
column_names = wine.feature_names
wine_data=pd.DataFrame(data=X ,columns=column_names)
wine_data.head()
```

```python
sample_wine_data=wine_data.sample(n=50,random_state=100)
```

```python
#sample mean
sample_mean=sample_wine_data["alcohol"].mean()
#sample_size
sample_size=len(sample_wine_data)
#std_dev
std_dev=sample_wine_data["alcohol"].std()
#Z critical value
z_critical=st.norm.ppf(.95)
# Margin of error = z_critical * (std_dev/sqrt(sample_size))
Margin_of_error = z_critical * (std_dev/sqrt(sample_size))
#confidence interval
confidance_interval=(sample_mean-Margin_of_error,sample_mean+Margin_of_error)
```

```python
print("Z critical value :",z_critical)
print("\nMargin of error :",Margin_of_error)
print("\nconfidance interval :",confidance_interval)
```
